# Hierarchical Indexing (MultiIndex)
    - from_tuples()
    - from_arrays()
    - from_product()
    - from_frame()

## from_tuples() 

```python
tuples = [
    (1, 'Python'),
    (1, 'SQL'),
    (2, 'Java'),
    (2, 'JavaScript'),
    (3, 'Csharp'),
]

data = {
    'Col1': [10, 20, 30, 40, 50],
    'Col2': [60, 70, 80, 90, 100]
}

mi = pd.MultiIndex.from_tuples(tuples, names=['Top', 'PLang'])

df = pd.DataFrame(data, index=mi)
```

## from_arrays()

```python
import numpy as np

arrays = [
    np.array(["A", "A", "B", "B", "C", "C", "D", "D"]),
    np.array(["one", "two", "one", "two", "one", "two", "one", "two"]),
    np.array(["1", "1", "2", "2", "3", "3", "4", "4"]),
]

data = {
    'col1': [1, 2, 3, 4, 5, 6, 7, 8], 
    'col2': [9, 10, 11, 12, 13, 14, 15, 16]
}

mi = pd.MultiIndex.from_arrays(arrays, names=['First', 'Second', 'Third'])

df_arrays = pd.DataFrame(data, index=mi)
```

## from_product()

```python
products = [
    ["A", "B", "C", "D"], 
    ["one", "two"]
]


data = {
    'col1': [1, 2, 3, 4, 5, 6, 7, 8], 
    'col2': [9, 10, 11, 12, 13, 14, 15, 16]
}

mi = pd.MultiIndex.from_product(products, names=['First', 'Second'])

df = pd.DataFrame(data, index=mi)
```

## from_frame()

```python
df = pd.DataFrame(
    {
        'First': ['A', 'A', 'B', 'B'],
        'Second': ['one', 'two', 'one', 'two']
    }
)

data = {'col1': [1, 2, 3, 4], 'col2': [5, 6, 7, 8]}

mi = pd.MultiIndex.from_frame(df)

df = pd.DataFrame(data, index=mi)
```

# Table Visualization

```python
import pandas as pd
import numpy as np

# create a sample dataframe
data = [
    [10, 15, 13, 17], 
    [21, 10, 14, 22]
]

index = pd.Index(['Group A', 'Group B'], name='Group')
columns = pd.MultiIndex.from_product([['Gold Points', 'Silver Points'], ['January', 'February']],
                                      names=['Type', 'Month'])
df = pd.DataFrame(data, index=index, columns=columns)

# create a table visualization with conditional formatting
def format_table(val):
    if val < 15:
        textcolor = "red"
        bgcolor = "pink"
        fontstyle = "italic"
    else:
        textcolor = "white"
        bgcolor = "teal"
        fontstyle = "bold"
    
    return f"background-color: {bgcolor}; color: {textcolor}; font-style: {fontstyle};"

# apply formatting funtion
table = df.style.applymap(format_table)
table.set_table_styles(
    [
    {'selector': 'th:hover', 'props': [('background-color', 'blue'), ('color', 'white')]},
    {'selector': 'td:hover', 'props': [('background-color', 'orange'), ('color', 'blue')]},
    {'selector': 'tr:hover', 'props': [('background-color', 'red'), ('color', 'white')]},
    ]
)

# display the table

table
```
